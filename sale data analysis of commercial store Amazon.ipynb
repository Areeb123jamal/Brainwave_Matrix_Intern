Data Cleaning

```python

df= df.fillna(0)
df.head()
```

```python
columns_to_drop = ['Purchase Order Number', 'Shipping Address Street 2', 'Group Name']
df = df.drop(columns=columns_to_drop)
```

```python
df['Total Charged'] = pd.to_numeric(df['Total Charged'], errors='coerce')
```

```python
df["Total Charged"] = df["Total Charged"].astype(float)
df.head()
```

```python
df['Total Charged'] = df['Total Charged'].astype(float)
df.describe()
```

 Descriptive Analysis

```python
total_spent = df["Total Charged"].sum()
print(f"Total spent: ${total_spent:.2f}")
```

```python
average_order_value = df["Total Charged"].mean()
print(f"Average order value: ${average_order_value:.2f}")
```

```python
median_order_value=df["Total Charged"].median()
print(f"median order value: ${median_order_value:.2f}")
```

```python
max_order_value=df["Total Charged"].max()
print(f"max order value: ${max_order_value:.2f}")
```

```python
min_order_value=df["Total Charged"].min()
print(f"min order value: ${min_order_value:.2f}")
```

How Much Have I Paid in Sales Tax?

```python
total_tax_paid = df["Tax Charged"].sum()
print(f"Total amount paid in taxes: ${total_tax_paid:.2f}")
```

```python
tax_rate = df["Tax Charged"].sum() / df["Total Charged"].sum()
print(f"Overall effective sales tax rate: {tax_rate:.2%}")
```

```python
total_orders = df["Order ID"].nunique()
print(f"Total number of orders: {total_orders}")
```

```python
payment_method_counts = df["Payment Instrument Type"].value_counts()
print(payment_method_counts)
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))["Total Charged"].sum()
print(monthly_sales)
```

```python
highest_sales_month = monthly_sales.idxmax()
lowest_sales_month = monthly_sales.idxmin()
print(f"Highest sales month: {highest_sales_month}, Sales: ${monthly_sales.max():.2f}")
print(f"Lowest sales month: {lowest_sales_month}, Sales: ${monthly_sales.min():.2f}")
```

```python
order_dates = df['Order Date'].sort_values()
time_between_orders = order_dates.diff().dropna()
average_time_between_orders = time_between_orders.mean()
print(f"Average time between orders: {average_time_between_orders}")
```

```python
order_status_counts = df["Order Status"].value_counts()
print(order_status_counts)
```

```python
shipping_charge_distribution = df["Shipping Charge"].value_counts()
print(shipping_charge_distribution)
```

```python
average_order_value_by_month = df.groupby(df['Order Date'].dt.to_period('M'))["Total Charged"].mean()
print(average_order_value_by_month)
```

```python
correlation = df["Subtotal"].corr(df["Tax Charged"])
print(f"Correlation between order size and tax charged: {correlation:.2f}")
```

Financial Analysis

```python
percentage_tax = (total_tax_paid / total_spent) * 100
print(f"Percentage of total spending that went to taxes: {percentage_tax:.2f}%")

```

```python
total_shipping_spent = df["Shipping Charge"].sum()
print(f"Total amount spent on shipping: ${total_shipping_spent:.2f}")
```

```python
returns_and_cancellations = df[df["Order Status"].isin(["Returned", "Canceled"])]
count_returns_cancellations = returns_and_cancellations["Order Status"].value_counts()

print(count_returns_cancellations)

total_returns_cancellations = count_returns_cancellations.sum()
print(f"Total number of returned and canceled orders: {total_returns_cancellations}")
```

```python
purchase_behavior = df.groupby(df['Order Date'].dt.to_period('M'))["Total Charged"].describe()
print(purchase_behavior)
```

Data Visualization

```python
import matplotlib.pyplot as plt
df.plot.bar(x='Order Date', y='Total Charged', rot=90)
plt.title("Total Charged per Order Date")
plt.xlabel("Order Date")
plt.ylabel("Total Charged ($)")
plt.show()  

```

```python
plt.figure(figsize=(20, 10))
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))

plt.title("Total Charged per Order Date")
plt.xlabel("Order Date")
plt.ylabel("Total Charged ($)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df["Total Charged"], bins=20, kde=True)
plt.title("Distribution of Total Order Costs")
plt.xlabel("Total Charged")
plt.ylabel("Frequency")
plt.show()
```

```python
df = df.sort_values(by="Order Date")
df["Cumulative Spending"] = df["Total Charged"].cumsum()
plt.figure(figsize=(10, 6))
plt.plot(df["Order Date"], df["Cumulative Spending"])
plt.title("Cumulative Spending Over Time")
plt.xlabel("Order Date")
plt.ylabel("Cumulative Spending")
plt.show()
```

```python
df["Tax Percentage"] = (df["Tax Charged"] / df["Total Charged"]) * 100
average_tax_percentage = df["Tax Percentage"].mean()

print(f"Average tax percentage per order: {average_tax_percentage:.2f}%")

plt.figure(figsize=(10, 6))
sns.histplot(df["Tax Percentage"], bins=20, kde=True)
plt.title("Distribution of Tax Percentage per Order")
plt.xlabel("Tax Percentage")
plt.ylabel("Frequency")
plt.show()
```

```python
monthly_spending = df.groupby(df["Order Date"].dt.to_period("M"))["Total Charged"].sum()

plt.figure(figsize=(10, 6))
monthly_spending.plot(kind="bar")
plt.title("Monthly Spending Patterns")
plt.xlabel("Month")
plt.ylabel("Total Charged")
plt.show()
```

```python
df["Shipping Percentage"] = (df["Shipping Charge"] / df["Total Charged"]) * 100
average_shipping_percentage = df["Shipping Percentage"].mean()

print(f"Average shipping cost as a percentage of total spending: {average_shipping_percentage:.2f}%")

plt.figure(figsize=(10, 6))
sns.histplot(df["Shipping Percentage"], bins=20, kde=True)
plt.title("Distribution of Shipping Cost Percentage per Order")
plt.xlabel("Shipping Percentage")
plt.ylabel("Frequency")
plt.show()
```
